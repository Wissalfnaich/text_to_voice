```python
# --- installation si nécessaire ---
# !pip install TTS soundfile

import soundfile as sf
from IPython.display import Audio, display
from TTS.api import TTS

# ====== PARAMÈTRES ======
VOICE_WAV = "Audio extrait.wav"   # <-- Ta voix (WAV, 10 à 40 sec conseillé)
OUTPUT_WAV = "resultat_tts.wav"
LANG = "fr"
MODEL = "tts_models/multilingual/multi-dataset/xtts_v2"

# ====== COLLE N'IMPORTE QUEL TEXTE ICI ======
TEXT = """
Coucou, Je suis super content de te revoir aujourd’hui. Tu sais, ton corps et ton cerveau font un travail incroyable, chaque jour.  
Mais parfois, c’est normal d’être fatiguée… ou d’avoir du mal à écouter tout en même temps. Tu veux que je t’explique pourquoi ?

Alors on va continuer ensemble. Un pas après l’autre. Une idée à la fois.  
Chaque petit effort compte vraiment ! Et moi, je crois en toi, à 100 % !

Si tu te sens perdue, pose des questions. Si tu es fière de toi, souris ! Et si tu veux recommencer, 
Bravo pour tout ce que tu fais déjà. Continue comme ça, tu es vraiment formidable !

"""

# ====== SYNTHÈSE VOCALE (CPU) ======
print("Chargement du modèle XTTS v2 (CPU)…")
tts = TTS(MODEL, gpu=False)  # <<< CPU UNIQUEMENT

print("Génération vocale…")
tts.tts_to_file(
    text=TEXT,
    speaker_wav=VOICE_WAV,
    language=LANG,
    file_path=OUTPUT_WAV,
    split_sentences=True  # Laisse XTTS gérer la respiration → pas de coupures
)

# ====== LECTURE DANS NOTEBOOK ======
audio, sr = sf.read(OUTPUT_WAV)
display(Audio(audio, rate=sr))
print("✅ Terminé →", OUTPUT_WAV)
```
